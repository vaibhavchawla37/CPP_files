```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
df = pd.read_csv('state_wise_daily.csv')
```

```python
df
```

```python
head=df['Status'].unique()
```

```python
head = np.insert(head,0,'Date')
```

```python
ndf=pd.DataFrame(columns=head)
```

```python
confirmed = df.iloc[0::3,]
recovered = df.iloc[1::3,]
deceased = df.iloc[2::3,]
```
